```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup


import os
from pandas.io.json import json_normalize
import urllib.request as request
import json

import operator
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
with request.urlopen("http://www.tanarang.com/english/raagIndex_eng.htm") as res:
    source_data = res.read()
```

```python
r_data = source_data.decode(encoding='UTF-8')
ragas_list = r_data.split("href")[19:139]
```

```python
raga_link = []
raga_name = []
for i in ragas_list:
    raga_link.append(i.split("</a>")[0].split(">")[0].split("=")[1].replace('"',""))
    raga_name.append(i.split("</a>")[0].split(">")[1])
```

```python
jati = []; thaat =[]; vadi_samvadi=[]; time=[]; mukhya_ang=[]; aaroh_avroh=[]; raag_url=[];ragaName=[]
for link,raga_n in zip(raga_link,raga_name):
    RAGA_URL = "http://www.tanarang.com/english/" + link
    raag_url.append(RAGA_URL)
    ragaName.append(raga_n)
    print(RAGA_URL)
    with request.urlopen(RAGA_URL) as res:
        #print(res)
        source_data = res.read()
   
    r_data = source_data.decode(encoding='UTF-8')
    #print(r_data.split("Jati</a></td>\r\n<td>"))
    if (r_data.find("Jati</a>") ==-1):
        jati.append("-")
    else:
        jati.append(r_data.split("Jati</a></td>\r\n<td")[1].split("</td></tr>")[0])
    if (r_data.find("Thaat</a>") ==-1):
        thaat.append("-")
    else:
        thaat.append(r_data.split("Thaat</a></td>\r\n<td")[1].split("</td></tr>")[0])
    if (r_data.find("Vadi/Samvadi</a>") ==-1):
        vadi_samvadi.append("-")
    else:
        vadi_samvadi.append(r_data.split("Vadi/Samvadi</a></td>\r\n<td")[1].split("</td></tr>")[0])
    if (r_data.find("Time</a>") ==-1):
        time.append("-")
    else:
        time.append(r_data.split("Time</a></td>\r\n<td")[1].split("</td></tr>")[0])
    if (r_data.find("Mukhya-Ang</a>") ==-1):
        mukhya_ang.append("-")
    else:
        mukhya_ang.append(r_data.split("Mukhya-Ang</a></td>\r\n<td")[1].split("</td></tr>")[0])
    if (r_data.find("Aaroh-Avroh</a>") ==-1):
        aaroh_avroh.append("-")
    else:
        aaroh_avroh.append(r_data.split("Aaroh-Avroh</a></td>\r\n<td")[1].split("</td></tr>")[0])
   
```

```python
av=[]
for i in aaroh_avroh:
    if (i.find("<strong>") == -1):
        av.append(i)
    else:
        if (i.find("</strong>") == -1):
            av.append(i.split("<strong>")[1].replace("'",""))
        else:
            av.append(i.split("<strong>")[1].split("</strong>")[0].replace("'",""))
```

```python
ma=[]
for i in mukhya_ang:
    if (i.find("<strong>") == -1):
        ma.append(i)
    else:
        if (i.find("</strong>") == -1):
            ma.append(i.split("<strong>")[1].replace("'","").replace("’",""))
        else:
            ma.append(i.split("<strong>")[1].split("</strong>")[0].replace("'","").replace("’",""))
```

```python
jati_ = []
for i in jati:
   jati_.append(i.split(">")[1].replace("Vakra", "").replace(" ",""))
```

```python
time_ =[]
for i in time:
    time_.append(i.split(">")[1])
```

```python
#1st Prahar of the Day , Second Prahar of the Night: Early Morning
#2nd Prahar of the day: Late morning
#3rd Prahar of the day: Afternoon
#4th Prahar of the day: Dusk

#1st Prahar of the night: Evening
#2nd Prahar of the Night : Night
#4th Prahar of the Night: Dawn
#3rd Prahar of the Night: Midnight
time__ = []

for i in time_:
    if (i.lower().find("1st prahar of the day") != -1):
        time__.append("Early Morning")
    
        
    elif (i.lower().find("2nd prahar of the day") != -1):
        time__.append("Late morning")
   
        
    elif (i.lower().find("3rd prahar of the day") != -1):
        time__.append("Afternoon")
   
        
    elif (i.lower().find("4th prahar of the day") != -1):
        time__.append("Dusk")
    

    elif (i.lower().find("1st prahar of the night") != -1):
        time__.append("Evening")
   
        
    elif (i.lower().find("2nd prahar of the night") != -1):
        time__.append("Night")
   
        
    elif (i.lower().find("3rd prahar of the night") != -1):
        time__.append("Dawn")
  
        
    elif (i.lower().find("4th prahar of the night") != -1):
        time__.append("Midnight")
        
        
    else:
        time__.append(i)
```

```python
thaat_ =[]
for i in thaat:
    if (i.find(" >") == -1):
        thaat_.append(i)
    else:
        if (i.find("(") == -1):
            thaat_.append(i.split(" >")[1])
        else:
            thaat_.append(i.split(" >")[1].split("(")[0])
```

```python
vadi_ = []; samvadi_ = []
for i in vadi_samvadi:
    if (i.find(" >") == -1):
        vadi_.append(i)
        samvadi_.append(i)
    else:
        vadi_.append(i.split(" >")[1].split("/")[0])
        samvadi_.append(i.split(" >")[1].split("/")[1])
        
```

```python
df= pd.DataFrame(list(zip(thaat_,vadi_,samvadi_,jati_,time__,ma,av,ragaName,raag_url)), columns =['Thaat','Vadi','Samvadi','Jati','Time of day','Mukhya-Ang','Aaroh-Avroh', 'Raag', 'link'])

```

### pre - processing

```python
data_update["Thaat"] = data_update["Thaat"].replace(to_replace =["-", "Not Defined"], 
                            value ="other")
data_update["Thaat"] = data_update["Thaat"].replace(to_replace =["Khamaj "], 
                            value ="Khamaj")
```

```python
data_update["Thaat"].unique()
```

```python
data_update["Vadi"] = data_update["Vadi"].replace(to_replace =["-"], 
                            value ="Other")
data_update["Vadi"] = data_update["Vadi"].replace(to_replace =["Tar Shadj"], 
                            value ="Shadj")
```

```python
data_update["Vadi"].unique()
```

```python
data_update["Samvadi"] = data_update["Samvadi"].replace(to_replace =["-"], 
                            value ="Other")
data_update["Samvadi"] = data_update["Samvadi"].replace(to_replace =["Shuddha Nishad"], 
                            value ="Nishad")
```

```python
data_update["Samvadi"].unique()
```

```python
data_update["Jati"] = data_update["Jati"].replace(to_replace =["ShadhavSampurna"], 
                            value ="Shadhav-Sampurna")
data_update["Jati"] = data_update["Jati"].replace(to_replace =["Audha-Audhav"], 
                            value ="Audhav-Audhav")
```

```python
data_update["Jati"].unique()
```

```python
data_update["Time of day"] = data_update["Time of day"].replace(to_replace =["Second Prahar of the Night or Anytime in Spring Season."], 
                            value ="Night")
data_update["Time of day"] = data_update["Time of day"].replace(to_replace =["Din ka Tritiya Prahar"], 
                            value ="Afternoon")
data_update["Time of day"] = data_update["Time of day"].replace(to_replace =["Sandhi-Prakash Raag"], 
                            value ="Dusk")#could be dawn or dusk, in this case it is dusk
data_update["Time of day"] = data_update["Time of day"].replace(to_replace =["(12 AM - 3 AM): 3 rd Prahar of the Night: Ratri ka Teesra Prahar"], 
                            value ="Midnight")
data_update["Time of day"] = data_update["Time of day"].replace(to_replace =["1sh Prahar of the day : Din ka Pratham Prahar", "(6 AM - 9 AM): Din ka Pratham Prahar "], 
                            value ="Early Morning")
data_update["Time of day"] = data_update["Time of day"].replace(to_replace =["Rainy Season - All times", "Anytime in Rainy season.",
       "Anytime in Rainy Season"], 
                            value ="Anytime")
```
